```python
def check_even_odd(num):
    if num % 2 == 0:
        print("Even Number")
    else:
        print("Odd Number")

check_even_odd(7)
```

```python
def square():
    n = int(input("Enter a number: "))
    print("Square =", n * n)

square()
```

```python
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact

print("Factorial =", factorial(5))
```

```python
def total_marks(marks):
    return sum(marks)

marks = [80, 75, 90]
print("Total =", total_marks(marks))
```
